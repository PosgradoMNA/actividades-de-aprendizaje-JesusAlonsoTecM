K-means es algo ingenuo: agrupa los datos en k grupos, incluso si k no es el número correcto de grupos a utilizar. Cuando hablamos de agrupamiento, es difícil saber cuántos grupos son óptimos... En nuestro conjunto de datos, ¿cuántos grupos son óptimos, es decir, tiene sentido, no queremos adivinar? Por lo tanto, cuando usamos el agrupamiento de k-medias, necesitamos una forma de determinar si estamos usando el número correcto de agrupamientos.

Un método para validar el número de CLUSTERS es el método del codo . La idea del método del codo es ejecutar el agrupamiento de k-medias en el conjunto de datos para un rango de valores de k (por ejemplo, k de 1 a 10), y para cada valor de k calcular la Suma de errores cuadrados (SSE).

Cuando K aumenta, los centroides están más cerca de los centroides de los grupos. Las mejoras disminuirán rápidamente en algún momento, creando la forma del codo. Ese es el valor óptimo para K.

```
from sklearn.cluster import KMeans

K_clusters = range(1,30)

kmeans = [KMeans(n_clusters=i) for i in K_clusters]

# Sabemos que no tienen nulos los datos, pues están preparados
Y = df[['LATITUD']]
X = df[['LONGITUD']]

score = [kmeans[i].fit(Y).score(Y) for i in range(len(kmeans))]

# Visualize
plt.plot(K_clusters, score)
plt.xlabel('Número de Clusters')
plt.ylabel('Score')
plt.title('Curva Elbow')
plt.show()
```

Al observar la gráfica, vemos que la gráfica se nivela lentamente después de 3 grupos. Esto implica que la adición de más clústeres no nos ayudará mucho, sin embargo, revisando los datos, 5 clusteres nos puede mantener al limite, así que tomaremos 5

```
# data
X1 = df[['LONGITUD', 'LATITUD']]

clusters = 5

# Clustering
kmean = KMeans(n_clusters=clusters)

# Prediciendo los clusters
labels = kmean.fit_predict(X1)
kmean.fit(X1)

# Centroides
centers = kmean.cluster_centers_

# Transformando a DataFrame para unir con los datos
df_cluster = pd.DataFrame(labels, columns=['Cluster'])

# Concatenando con los valores iniciales
X2 = pd.concat([X1, df_cluster], axis = 1)
 
X2.head()
```

```
# Para poder graficar la variable SEMAFORO, debemos codificarla a numérica
from sklearn.preprocessing import OrdinalEncoder

# Creamos el codificador indicandole el orden de la variables
encoder = OrdinalEncoder(categories=[["Rojo", 
                                      "Amarillo", 
                                      "Verde"]])

# Ajustamos el codificador con la variable semaforo y la transformamos
encoder.fit(df[["SEMAFORO"]])
df["SEMAFORO-encoded"] = encoder.transform(df[["SEMAFORO"]])

# mostramos los valores de la variable antes y después de aplicar la codificación ordinal.
df[["SEMAFORO", "SEMAFORO-encoded"]].head(100)
```

```
#Graficar los nodos dados (ubicaciones)
Lat=df['LATITUD']
Lon=df['LONGITUD']

X = []
for i in range(len(df['LATITUD'])):
    X.append(Lat[i])
    X.append(Lon[i])

X = np.array(X)
X = X.reshape(-1, 2,)

plt.scatter(Lat, Lon)
plt.show()
```

Dentro de nuestro dataset utilizaremos la variable SEMAFORO ya codificada a numérica para darle una relacionada de peso de cada nodo. Pues bien, vamos a considerarla al ejecutar el algoritmo K-Means.

```
# Habiendo pasado la variable de SEMAFORO a numéricas, le damos peso ahora si y los pasamos por K-MEANS

# Transformaremos ahora los valores actuales de SEMAFORO para poderlos gráficar
Peso = df['SEMAFORO-encoded']

KMean = KMeans(n_clusters=clusters)
KMean_g = KMean.fit_predict(X)
KMean.fit(X, sample_weight=Peso)
```

Lo siguiente que haremos será establecer los centroides de cada clúster, que ahora sí serán Centros de Gravedad; es decir, la ubicación de nuestros contendores. En primer lugar, el modelo nos dará las coordenadas. Utilizaremos la siguiente línea de código:

```
# Centroides de los clusters definidos 
KMean.cluster_centers_
```

Los centroides han cambiado, ahora son centros de gravedad afectados por el peso de cada nodo. Es posible que incluso haya cambiado la agrupación de los nodos (composición de los clusters).

```
# Graficar todas las coordenadas (Puntos y centroides)
plt.scatter(X[:,0], X[:,1], c=KMean_g) # Puntos iniciales

# Centroides
plt.scatter(KMean.cluster_centers_[0][0], KMean.cluster_centers_[0][1], s=50, c='b')
plt.scatter(KMean.cluster_centers_[1][0], KMean.cluster_centers_[1][1], s=50, c='b')
plt.scatter(KMean.cluster_centers_[2][0], KMean.cluster_centers_[2][1], s=50, c='b')
plt.scatter(KMean.cluster_centers_[3][0], KMean.cluster_centers_[3][1], s=50, c='b')
plt.scatter(KMean.cluster_centers_[4][0], KMean.cluster_centers_[4][1], s=50, c='b')
```

Con este proceso de graficación, ahora tenemos un modelo capaz de agrupar nuestros nodos en clusters de acuerdo a atributos geoespaciales; capaz de determinar los centros de gravedad de cada cluster de acuerdo a atributos no espaciales **(en nuestro caso el peso de la variable SEMAFORO que nos indica la calidad buena o mala del agua);** capaz de complementarse con una capa de visualización de mapa de calor que nos permite apreciar la densidad y todo esto puede visualizarse en un entorno geográfico real.

```
import folium
import statistics
from folium.plugins import HeatMap
import geopy
from geopy.extra.rate_limiter import RateLimiter

# Calcular la media de las latitudes y las longitudes para centrar el mapa 
mediaLong = statistics.mean(Lon) 
mediaLat = statistics.mean(Lat) 

# Crear un objeto de mapa base Map() 
mapa = folium.Map(location=[mediaLat, mediaLong], zoom_start = 12) 

# Crear una capa de mapa de calor 
mapa_calor = HeatMap( list(zip(Lat, Lon, df['SEMAFORO-encoded'])), 
                     min_opacity=0.2, 
                     radius=50, 
                     blur=50, 
                     max_zoom=1)

#Creamos el marcador de Centro de Gravedad
tooltip = 'Centro de gravedad'
folium.Marker([KMean.cluster_centers_[0][0], KMean.cluster_centers_[0][1]], popup="Centro", tooltip = tooltip).add_to(mapa)
folium.Marker([KMean.cluster_centers_[1][0], KMean.cluster_centers_[1][1]], popup="Centro", tooltip = tooltip).add_to(mapa)
folium.Marker([KMean.cluster_centers_[2][0], KMean.cluster_centers_[2][1]], popup="Centro", tooltip = tooltip).add_to(mapa)
folium.Marker([KMean.cluster_centers_[3][0], KMean.cluster_centers_[3][1]], popup="Centro", tooltip = tooltip).add_to(mapa)
folium.Marker([KMean.cluster_centers_[4][0], KMean.cluster_centers_[4][1]], popup="Centro", tooltip = tooltip).add_to(mapa)

# Adherimos la capa de mapa de calor al mapa principal 
mapa_calor.add_to(mapa) 
mapa
```
